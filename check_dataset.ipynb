# ECGデータの解析

2023/05/03

BioSPPyなる生体信号解析ライブラリを使う

参考：https://zenn.dev/kn1cht/articles/apple-watch-ecg-python

とりあえず，睡眠ステージ判定前30秒のデータに対して処理する．

下記は，最初の判定（Wake：72000サンプル）における処理

```python
from biosppy.signals import ecg

stage_sample = 4368000#72000
data_num = 6000

ecg_data = ecg.ecg(signal=ecg_sig[stage_sample-data_num-600:stage_sample+600], sampling_rate=200., show=True, interactive=False)#66000~72000までのデータ
```

RRIntervalの検出を行う．

```python
(ts, filtered, rpeaks, templates_ts, templates, heart_rate_ts, heart_rate) = ecg_data
#rpeaks_christov, = ecg.christov_segmenter(filtered, 200.) #一番検出精度がいいらしい
rpeaks_engzee, = ecg.engzee_segmenter(filtered, 200.) #こっちの方がロバストだった
```

```python
ts_peaks = ts[rpeaks_engzee]
rri = np.diff(ts_peaks) * 1000
plt.plot(ts_peaks[0:-1], rri)
plt.scatter(ts_peaks[0:-1], rri)
plt.ylabel("Raw_RRI [ms]")
plt.xlabel("Time [s]")
plt.xlim([0, 40])
plt.ylim([0, 1500])
plt.grid()
```

サンプリングレートがバラバラなので，リサンプリングする．

スプライン補完を実行する．

```python
import scipy

spline_func = scipy.interpolate.interp1d(ts_peaks[:-1], rri, kind='linear', bounds_error=False, fill_value='extrapolate')
ts_1sec = np.arange(0, 36, 1)
rri_1sec = spline_func(ts_1sec)
plt.plot(ts_1sec, rri_1sec)
plt.scatter(ts_1sec, rri_1sec)
plt.ylabel("Interpolated RRI [ms]")
plt.xlabel("Time [s]")
#plt.xlim([0, 30])
```

比較

```python
plt.plot(ts_peaks[0:-1], rri)
plt.scatter(ts_peaks[0:-1], rri, label="Raw")

plt.plot(ts_1sec, rri_1sec, linestyle="--")
plt.scatter(ts_1sec, rri_1sec, marker="x", label="Interpolated")
plt.ylabel("RRI [ms]")
plt.xlabel("Time [s]")
plt.legend()
```
