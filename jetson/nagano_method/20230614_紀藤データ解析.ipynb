```python
import pandas as pd
import openpyxl
import matplotlib.pyplot as plt
import numpy as np

#データ読み込み
data = pd.read_csv("Kito2.csv")
data
```

```python
lag_win = 60 #ラグ特徴量のウィンドウ幅
df_lag = pd.DataFrame()
for i in range(1, lag_win):
    df_lag['x_lag{}'.format(i)] = data['PPG'].shift(i)
df_lag
train_df = pd.concat([data['PPG'], df_lag], axis=1)
train_df.dropna(inplace=True)
train_df.reset_index(inplace=True)
train_df
```

```python
def outlier_2s(df):

    for i in range(len(df.columns)):

        # 列を抽出する
        col = df.iloc[:,i]

        # 平均と標準偏差
        average = np.mean(col)
        sd = np.std(col)

        # 外れ値の基準点
        outlier_min = average - (sd) * 2
        outlier_max = average + (sd) * 2

        # 範囲から外れている値を除く
        col[col < outlier_min] = None
        col[col > outlier_max] = None

    return df
```

```python
df2 = outlier_2s(data)
df2
```

```python
def movingAverage(data, num):
    b=np.ones(num)/num
    filtered = np.convolve(data, b, mode='same')#移動平均
    return filtered
```
